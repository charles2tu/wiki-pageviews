```python
from statsforecast import StatsForecast
from statsforecast.models import ARIMA
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import us # for state names
from scipy import signal
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from scipy import signal
import warnings
warnings.filterwarnings("ignore")
```

```python
# hospitalization data

hosp = merged_df
hosp = hosp[hosp['date'] <= '2024-04-27'] # end date for 2023-2024 flu season
cutoff = '2022-10-01' # cutoff between train and test sets
hosp_train = hosp[hosp['date'] <= cutoff]

# exogenous data
exog = hosp[['date', 'Influenza A virus']]
exog = exog[exog['date'] >= hosp['date'].min()]
exog = exog[exog['date'] <= hosp['date'].max()]
# exog = exog.reset_index(drop = True)

# all locations
geo = pd.Series(i for i in merged_df.columns[10:-1,])
geo = geo.str.replace('hosp_', '', regex=True)
geo
model = ARIMA(
    order=(1, 1, 1),           # ARIMA(1,1,1)
    include_mean=False,        # No intercept if the series is stationary
    include_drift=False,       # No drift component
    season_length=1,           # No seasonal effects
    seasonal_order=(0, 0, 0)   # Explicitly stating no seasonality
)
```

```python


# keeping only variables highly correlated with target
def filter_corr(state, data, hosp_state, percentile_value=25): 
    # print(f"Filtering correlations for state: {state}")
    # print(f"data_sub shape: {data.shape}, hosp_sub shape: {hosp.shape}")
    # print(f"data_sub columns: {data.columns.tolist()}")
    # print(f"hosp_sub sample:\n{hosp.head()}")
    hosp_sub = hosp_state[:hosp_train.shape[0]]
    data_sub = data.iloc[:hosp_train.shape[0]]
    columns_to_keep = []
    correlations = data_sub.corrwith(hosp_sub)
    
    # percentile-based threshold based on correlation distribution
    corr_threshold = np.percentile(correlations.dropna(), percentile_value)
    
    filtered_corrs = correlations[correlations >= corr_threshold]
    selected_columns = filtered_corrs.index.tolist()
    # computing the correlation matrix for the selected columns
    selected_corr_matrix = data_sub[selected_columns].corr().abs()
    
    # identifying highly correlated pairs (above 0.99) to avoid multicollinearity
    to_remove = set()
    for i in range(len(selected_columns)):
        for j in range(i + 1, len(selected_columns)):
            if selected_corr_matrix.iloc[i, j] > 0.90:
                # keeping only one of the two variables
                to_remove.add(selected_columns[j])  # removing the second one

                
    final_columns = [col for col in selected_columns if col not in to_remove]
    corrs_35 = filtered_corrs.loc[final_columns].nlargest(35)
    final_35 = corrs_35.index.tolist()

    return final_35, filtered_corrs.loc[final_35], corr_threshold

# getting the state name corresponding to its code
def get_state_name(state):
    state_code = state.split('-')[1] if '-' in state else state

    # Special case for 'US-DC'
    if state_code == 'DC':
        state_name = 'District of Columbia'
    else:
        state_name = us.states.lookup(state_code).name if us.states.lookup(state_code) else 'US'

    return state_name

# preparing the data in the format of the statsforecast package
def data_prep(state, df):
    state_code = state.split('-')[1] if '-' in state else state
    print(state_code)
    state_name = get_state_name(state)
    print(state_name)

    # exogenous
    # data = df[df.columns[df.columns.str.contains(state+'_')]]
    
    # hospitalizations
    hosp_state = hosp[f'hosp_{state_name}']
    cols, corrs, threshold = filter_corr(state, df, hosp_state)
    print('corr threshold:', threshold)
    print('num preds:', corrs.shape[0])

    # matching format of statsforecast
    Y_ts = pd.DataFrame(hosp_state)
    Y_ts = Y_ts.rename(columns={hosp_state.name: 'y'})
    Y_ts.insert(0, 'ds', hosp['date'])
    Y_ts["ds"] = pd.to_datetime(Y_ts["ds"])
    Y_ts.insert(0, 'unique_id', '1')
    
    X_ts = pd.DataFrame()
    X_ts['Influenza A virus'] = df['Influenza A virus']    
    X_ts.insert(0, 'ds', df['date']) # adding date from format of statsforecast
    X_ts.insert(0, 'unique_id', '1')
    X_ts["ds"] = pd.to_datetime(X_ts["ds"])

    return X_ts, Y_ts, hosp_state, corrs, threshold

# ARIMAX forecast for one state
start = 104 # using rolling window of past 104 weeks
min_date = '2022-10-17'
def arima_forecast(X_ts, Y_ts):
    print(X_ts)
    print(Y_ts)
    predictions = []
    time_steps = []
    dates = X_ts['ds'].unique()
    dates = dates[dates > np.datetime64('2019-01-01')] # starts the test set on 2021-01-01

    for n in range(start, len(dates)):
        dtrain = dates[n-104:n] # rolling window of 104 weeks
        dtest = dates[n]

        Y_train = Y_ts.query('ds in @dtrain')
        Y_test = Y_ts.query('ds == @dtest')

        X_train = X_ts.query('ds in @dtrain')
        X_test = X_ts.query('ds == @dtest')

        # merging training data
        train = Y_train.merge(X_train, how='left', on=['unique_id', 'ds'])

        # creating a StatsForecast instance with weekly frequency (hospitalization date corresponds to ending Saturday)
        sf = StatsForecast(models=[model], freq='W-SAT', n_jobs=-1)
        # print(train)
        # forecasting the next week
        fcst = sf.forecast(df=train, h=1, X_df=X_test)
        predictions.append(fcst['ARIMA'].values[0])  # appending the forecasted value
        time_steps.append(fcst['ds'].values[0])
        
    truth = Y_ts[Y_ts['ds'] >= min_date]
    preds = pd.DataFrame({'time_steps': time_steps, 'predictions': predictions})
    preds = preds[preds['time_steps'] >= min_date]
    preds = preds.reset_index(drop=True)

    # computing RMSE and MAE
    rmse = np.sqrt(mean_squared_error(preds['predictions'], truth.iloc[:,2]))
    mae = mean_absolute_error(preds['predictions'], truth.iloc[:,2])
    print('rmse:', rmse)
    print('mae:', mae)
    
    return preds['predictions'], preds['time_steps'], rmse, mae


# ARIMA forecast for one state without exogenous variables
def arima_noexog(Y_ts):

    predictions = []
    time_steps = []
    dates = X_ts['ds'].unique()
    # print(dates)
    dates = dates[dates > np.datetime64('2019-01-01')] # starts the test set on 2021-01-01

    for n in range(start, len(dates)):
        dtrain = dates[n-104:n] # rolling window of 104 weeks
        dtest = dates[n]

        Y_train = Y_ts.query('ds in @dtrain')
        Y_test = Y_ts.query('ds == @dtest')

        X_train = X_ts.query('ds in @dtrain')
        X_test = X_ts.query('ds == @dtest')

        # merging training data
        train = Y_train.merge(X_train, how='left', on=['unique_id', 'ds'])

        # creating a StatsForecast instance with weekly frequency and configured ARIMA (hospitalization date corresponds to ending Saturday)
        sf = StatsForecast(models=[model], freq='W-SAT', n_jobs=-1)

        # forecasting the next week
        fcst = sf.forecast(df=Y_train, h=1)  # no exogenous variables
        # print(fcst['ds'])
        # print(min_date)
        predictions.append(fcst['ARIMA'].values[0])  # appending the forecasted value
        time_steps.append(fcst['ds'].values[0])

    truth = Y_ts[Y_ts['ds'] >= min_date]
    preds = pd.DataFrame({'time_steps': time_steps, 'predictions': predictions})
    preds = preds[preds['time_steps'] >= min_date]
    preds = preds.reset_index(drop=True)
    # print(preds['predictions'])
    # computing RMSE and MAE
    rmse = np.sqrt(mean_squared_error(preds['predictions'], truth.iloc[:,2]))
    mae = mean_absolute_error(preds['predictions'], truth.iloc[:,2])
    print('rmse:', rmse)
    print('mae:', mae)

    return preds['predictions'], preds['time_steps'], rmse, mae

```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.seasonal import seasonal_decompose

# Ensure datetime index
merged_df.index = pd.to_datetime(merged_df.index)

# Define your target
y = merged_df["hosp_US"]

# Seasonal decomposition (Additive to handle zeros)
decompose_result = seasonal_decompose(y, model="additive", period=52)
residual = decompose_result.resid.dropna()

# Containers for direct forecasts
direct_forecasts = []
forecast_steps = 208  # 4 years ahead weekly

# Direct forecasting loop
for step in range(1, forecast_steps + 1):
    print(f"Forecasting step {step}/{forecast_steps}")
    # Shift residual to directly forecast 'step' ahead
    shifted_residual = residual.shift(-step).dropna()

    # Fit ARIMA(1,1,1) model
    model = ARIMA(residual.iloc[:-step], order=(1,1,1))
    model_fit = model.fit()

    # Forecast next value explicitly
    forecast = model_fit.forecast(steps=1)

    # Append forecast
    direct_forecasts.append(forecast.iloc[0])

# Reconstruct forecast with seasonal and trend components
seasonal_pattern = np.tile(decompose_result.seasonal[-52:], 4)
trend_mean = decompose_result.trend.dropna().iloc[-52:].mean()

forecast_full = np.array(direct_forecasts) + seasonal_pattern + trend_mean
forecast_final = np.maximum(forecast_full, 0)

# Forecast dates
forecast_index = pd.date_range(start=merged_df.index[-1] + pd.Timedelta(weeks=1), periods=forecast_steps, freq='W')

# Plot the forecast
plt.figure(figsize=(14,7))
plt.plot(merged_df.index, y, label="Actual Hospital Admissions")
plt.plot(forecast_index, forecast_final, linestyle='--', color='green', label="Direct ARIMA(1,1,1) Forecast")
plt.xlabel("Date")
plt.ylabel("Hospital Admissions")
plt.title("4-Year Direct ARIMA(1,1,1) Forecast")
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_absolute_error, mean_squared_error
from statsmodels.tsa.seasonal import seasonal_decompose

# Ensure datetime index
merged_df.index = pd.to_datetime(merged_df.index)

# Define data
x = merged_df["Influenza A virus"]
y = merged_df["hosp_US"]

# Split data into train and test
train_size = int(len(y) * 0.8)
y_train, y_test = y[:train_size], y[train_size:]
x_train, x_test = x[:train_size], x[train_size:]

# Seasonal decomposition
result = seasonal_decompose(y_train, model="additive", period=52)
residual_train = result.resid.dropna()
x_train_aligned = x_train.loc[residual_train.index]

# Direct Forecasting
arima_forecasts = []
arimax_forecasts = []

forecast_horizon = len(y_test)

for step in range(1, forecast_horizon + 1):
    # ARIMA model
    model_arima = SARIMAX(y_train, order=(1,1,1))
    model_fit_arima = model_arima.fit(disp=False)
    forecast_arima = model_fit_arima.forecast(steps=step).iloc[-1]
    forecast_arima = max(forecast_arima, 0)
    arima_forecasts.append(forecast_arima)

    # ARIMAX model
    model_arimax = SARIMAX(y_train, exog=x_train, order=(1,1,1))
    model_fit_arimax = model_arimax.fit(disp=False)

    forecast_arimax = model_fit_arimax.forecast(steps=1, exog=x_test.iloc[[step-1]])[0]
    forecast_arimax = max(forecast_arimax, 0)
    arimax_forecasts.append(forecast_arimax)

# Evaluate performance
mae_arima = mean_absolute_error(y_test, arima_forecasts)
rmse_arima = np.sqrt(mean_squared_error(y_test, arima_forecasts))

mae_arimax = mean_absolute_error(y_test, arimax_forecasts)
rmse_arimax = np.sqrt(mean_squared_error(y_test, arimax_forecasts))

print("ARIMA MAE:", mae_arima)
print("ARIMA RMSE:", rmse_arima)

print("ARIMAX MAE:", mae_arimax)
print("ARIMAX RMSE:", rmse_arimax)

# Visual Comparison
plt.figure(figsize=(14,7))
plt.plot(y_test.index, y_test, label="Actual Hospital Admissions", color="black")
plt.plot(y_test.index, arima_forecasts, linestyle="--", color="blue", label="ARIMA Forecast")
plt.plot(y_test.index, arimax_forecasts, linestyle="--", color="red", label="ARIMAX Forecast (Wikipedia Views)")
plt.xlabel("Date")
plt.ylabel("Hospital Admissions")
plt.title("ARIMA vs ARIMAX Model Forecast Comparison")
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```
